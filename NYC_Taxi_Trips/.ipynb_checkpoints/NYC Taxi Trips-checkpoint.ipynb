#  Viajes de Taxi NYC - Análisis de Series Temporales

Este notebook presenta un análisis exhaustivo de los datos de viajes de taxis amarillos de Nueva York correspondientes al año 2016. 

El objetivo principal de este análisis es:
- Explorar y limpiar el dataset para garantizar la calidad de los datos
- Identificar patrones temporales en la demanda de taxis
- Analizar la duración de los viajes y su relación con diferentes variables
- Extraer insights valiosos para la optimización de servicios de transporte

A través de técnicas de análisis de series temporales y visualización de datos, buscaremos comprender cómo varía la demanda de taxis a lo largo del día, días de la semana y meses del año, proporcionando una base sólida para la toma de decisiones estratégicas en el sector de transporte.

**Referencia del Dataset:**  
Meg Risdal. *New York City Taxi Trip Duration*. Kaggle, 2017. [https://kaggle.com/competitions/nyc-taxi-trip-duration](https://kaggle.com/competitions/nyc-taxi-trip-duration)

## Diccionario de datos

| **Variable**            | **Descripción**                                                                 |
|--------------------------|--------------------------------------------------------------------------------|
| **id**                  | Identificador único de cada viaje                                              |
| **vendor_id**           | Código que indica el proveedor asociado al registro del viaje                  |
| **pickup_datetime**     | Fecha y hora en que se activó el taxímetro                                     |
| **dropoff_datetime**    | Fecha y hora en que se desactivó el taxímetro                                  |
| **passenger_count**     | Número de pasajeros en el vehículo (valor ingresado por el conductor)          |
| **pickup_longitude**    | Longitud donde se activó el taxímetro                                          |
| **pickup_latitude**     | Latitud donde se activó el taxímetro                                           |
| **dropoff_longitude**   | Longitud donde se desactivó el taxímetro                                       |
| **dropoff_latitude**    | Latitud donde se desactivó el taxímetro                                        |
| **store_and_fwd_flag**  | Indica si el registro del viaje fue almacenado en la memoria del vehículo antes de enviarse al proveedor (Y/N) |
| **trip_duration**       | Duración del viaje en segundos                                                 |


# 1. Data Loading 
Primero, se cargan los datos y se hace una revision inicial del dataframe para analizar y decidir los primeros pasos a realizar para tener un dataframe de calidad. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Configuración de visualización
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = (12, 6)
sns.set_palette("husl")
```

```python
# Read all CSV files in the folder and merge them into a DataFrame
df = pd.concat(map(pd.read_csv, glob.glob("NYC_Taxi_Trip_Duration/*.csv")), ignore_index=True)
```

```python
# Display Dataframe
display(df)
```

```python
# Column info and data types
df.info()
```

**Primeros pasos**   
Las columnas de fecha deben ser convertidas al tipo `datetime` y deben extraerse características (hora, día, día de la semana, mes). Los valores nulos deben ser revisados y los datos inválidos deben ser limpiados (por ejemplo, coordenadas, duraciones negativas o en cero). Las distribuciones de variables clave (duración del viaje, número de pasajeros, ubicaciones de inicio y destino) deben ser exploradas.

```python
# Descriptive statistics
df.describe(include="all").transpose()
```

Se observa que algunas columnas contienen datos faltantes, valores atípicos o rangos muy amplios, como es el caso de la duración del viaje y las coordenadas de pickup y dropoff, lo que indica que es necesario revisarlas y tratarlas antes de continuar con el proceso de modelado.

# 2. Data Cleaning 

Primer se buscaran duplicados. 

```python
# Show duplicate rows
print("Number of duplicate rows:", df.duplicated().sum())
```

No hay duplicados, ahora, se calculara el porcentaje de NaNs en el dataset. 

```python
# Percentage of missing values per column
null_percent = (df.isnull().sum() / len(df)) * 100
print("Porcentaje de valores nulos por columna:")
print(null_percent)

# Delete NANs
critical_columns = ['id', 'pickup_datetime', 'pickup_longitude', 'pickup_latitude', 'trip_duration']
df_clean = df.dropna(subset=critical_columns).copy()
print(f"Filas después de eliminar nulos en columnas críticas: {len(df_clean)}")
```

```python
# Manejo específico por columna
if df_clean['vendor_id'].isnull().any():
    df_clean['vendor_id'] = df_clean['vendor_id'].fillna(df_clean['vendor_id'].mode()[0])

if df_clean['passenger_count'].isnull().any():
    df_clean['passenger_count'] = df_clean['passenger_count'].fillna(1)
```

```python
# Variable categórica
df_clean['store_and_fwd_flag'] = df_clean['store_and_fwd_flag'].fillna('N')
```

```python
# Calcular dropoff_datetime si falta
mask = df_clean['dropoff_datetime'].isnull() & df_clean['pickup_datetime'].notnull() & df_clean['trip_duration'].notnull()
df_clean.loc[mask, 'dropoff_datetime'] = pd.to_datetime(df_clean.loc[mask, 'pickup_datetime']) + pd.to_timedelta(df_clean.loc[mask, 'trip_duration'], unit='s')
```

```python
# Coordenadas de dropoff que faltan
dropoff_coords_null = df_clean['dropoff_longitude'].isnull() | df_clean['dropoff_latitude'].isnull()
df_clean.loc[dropoff_coords_null, 'dropoff_longitude'] = df_clean.loc[dropoff_coords_null, 'pickup_longitude']
df_clean.loc[dropoff_coords_null, 'dropoff_latitude'] = df_clean.loc[dropoff_coords_null, 'pickup_latitude']
```

**Conversión de tipos de datos**

```python
# Convertir columnas de fecha
df_clean['pickup_datetime'] = pd.to_datetime(df_clean['pickup_datetime'])
df_clean['dropoff_datetime'] = pd.to_datetime(df_clean['dropoff_datetime'])

# Convertir tipos numéricos
df_clean['vendor_id'] = df_clean['vendor_id'].astype(int)
df_clean['passenger_count'] = df_clean['passenger_count'].astype(int)
df_clean['trip_duration'] = df_clean['trip_duration'].astype(int)

# Convertir booleanos
df_clean['store_and_fwd_flag'] = df_clean['store_and_fwd_flag'].map({'Y': True, 'N': False, 'y': True, 'n': False})

print("Tipos de datos después de la conversión:")
print(df_clean.dtypes)
```

```python
# VALIDACIÓN Y FILTRADO FINAL

# Verificar que los datos tengan sentido
negative_duration = (df_clean['trip_duration'] <= 0).sum()
print(f"Viajes con duración <= 0: {negative_duration}")

invalid_dates = (df_clean['dropoff_datetime'] < df_clean['pickup_datetime']).sum()
print(f"Viajes con dropoff antes de pickup: {invalid_dates}")

# Coordenadas dentro de rangos razonables para NYC
nyc_bounds = {
    'min_lon': -74.05, 'max_lon': -73.70,
    'min_lat': 40.50, 'max_lat': 40.90
}

outside_nyc = (
    (~df_clean['pickup_longitude'].between(nyc_bounds['min_lon'], nyc_bounds['max_lon'])) |
    (~df_clean['pickup_latitude'].between(nyc_bounds['min_lat'], nyc_bounds['max_lat']))
).sum()
print(f"Viajes con coordenadas fuera de NYC: {outside_nyc}")
```

```python
# Limpiar casos problemáticos
df_final = df_clean[
    (df_clean['trip_duration'] > 0) &
    (df_clean['dropoff_datetime'] >= df_clean['pickup_datetime']) &
    (df_clean['pickup_longitude'].between(nyc_bounds['min_lon'], nyc_bounds['max_lon'])) &
    (df_clean['pickup_latitude'].between(nyc_bounds['min_lat'], nyc_bounds['max_lat']))
].copy()

print(f"\nDataset final: {len(df_final)} filas")
print(f"Valores nulos restantes: {df_final.isnull().sum().sum()}")
```

# 3. Preparación para análisis temporal

```python
# Extraer características temporales
df_final['pickup_hour'] = df_final['pickup_datetime'].dt.hour
df_final['pickup_day'] = df_final['pickup_datetime'].dt.day
df_final['pickup_dayofweek'] = df_final['pickup_datetime'].dt.dayofweek
df_final['pickup_month'] = df_final['pickup_datetime'].dt.month
df_final['pickup_week'] = df_final['pickup_datetime'].dt.isocalendar().week
```

```python
# Crear variable para día de la semana (nombre)
df_final['pickup_weekday_name'] = df_final['pickup_datetime'].dt.day_name()

# Convertir duración a minutos para mejor interpretación
df_final['trip_duration_min'] = df_final['trip_duration'] / 60

# Clasificar días en laborables vs fin de semana
df_final['is_weekend'] = df_final['pickup_dayofweek'].isin([5, 6])  # 5=Sábado, 6=Domingo

print("Características temporales creadas:\n")
display(df_final[['pickup_datetime', 'pickup_hour', 'pickup_day', 'pickup_dayofweek', 
                'pickup_month', 'pickup_weekday_name', 'is_weekend']].head())
```

# 4. Análisis de patrones
## Análisis de patrones horarios
### Viajes por hora del día

```python
# Viajes por hora del día
hourly_trips = df_final.groupby('pickup_hour').size()
hourly_duration = df_final.groupby('pickup_hour')['trip_duration_min'].mean()
hourly_passengers = df_final.groupby('pickup_hour')['passenger_count'].mean()

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.plot(hourly_trips.index, hourly_trips.values, marker='o', linewidth=2, color='cornflowerblue')
plt.title('Número de Viajes por Hora del Día')
plt.xlabel('Hora del Día')
plt.ylabel('Número de Viajes')
plt.grid(True, alpha=0.3)
plt.xticks(range(0, 24, 2))

plt.subplot(2, 2, 2)
plt.plot(hourly_duration.index, hourly_duration.values, marker='o', color='mediumslateblue', linewidth=2)
plt.title('Duración Promedio de Viajes por Hora')
plt.xlabel('Hora del Día')
plt.ylabel('Duración Promedio (min)')
plt.grid(True, alpha=0.3)
plt.xticks(range(0, 24, 2))

plt.subplot(2, 2, 3)
plt.plot(hourly_passengers.index, hourly_passengers.values, marker='o', color='deeppink', linewidth=2)
plt.title('Pasajeros Promedio por Hora')
plt.xlabel('Hora del Día')
plt.ylabel('Pasajeros Promedio')
plt.grid(True, alpha=0.3)
plt.xticks(range(0, 24, 2))

plt.tight_layout()
plt.show()

print(f"Hora pico de viajes: {hourly_trips.idxmax()}:00 hrs ({hourly_trips.max():,.0f} viajes)")
print(f"Hora con mayor duración promedio: {hourly_duration.idxmax()}:00 hrs ({hourly_duration.max():.1f} min)")
print(f"Hora con más pasajeros promedio: {hourly_passengers.idxmax()}:00 hrs ({hourly_passengers.max():.2f} pasajeros)")
```

### Viajes por día de la semana

```python
weekday_trips = df_final.groupby('pickup_dayofweek').size()
weekday_duration = df_final.groupby('pickup_dayofweek')['trip_duration_min'].mean()
weekday_names = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.bar(weekday_names, weekday_trips.values, color='cornflowerblue', alpha=0.7)
plt.title('Número de Viajes por Día de la Semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Número de Viajes')
plt.xticks(rotation=45)
for i, v in enumerate(weekday_trips.values):
    plt.text(i, v, f'{v:,}', ha='center', va='bottom')

plt.subplot(2, 2, 2)
plt.bar(weekday_names, weekday_duration.values, color='deeppink', alpha=0.7)
plt.title('Duración Promedio por Día de la Semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Duración Promedio (min)')
plt.xticks(rotation=45)
for i, v in enumerate(weekday_duration.values):
    plt.text(i, v, f'{v:.1f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
    
print(f"Día con más viajes: {weekday_names[weekday_trips.idxmax()]} ({weekday_trips.max():,.0f} viajes)")
print(f"Día con menos viajes: {weekday_names[weekday_trips.idxmin()]} ({weekday_trips.min():,.0f} viajes)")
```

### Viajes por mes

```python
# Viajes por mes
monthly_trips = df_final.groupby('pickup_month').size()
month_names = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']

plt.subplot(2, 2, 3)
plt.bar(month_names[:len(monthly_trips)], monthly_trips.values, color='mediumslateblue', alpha=0.7)
plt.title('Número de Viajes por Mes')
plt.xlabel('Mes')
plt.ylabel('Número de Viajes')
for i, v in enumerate(monthly_trips.values):
    plt.text(i, v, f'{v:,}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## Análisis de tendencia temporal

```python
# Agrupar por fecha para ver tendencia
daily_trips = df_final.groupby(df_final['pickup_datetime'].dt.date).agg({
    'id': 'count',
    'trip_duration_min': 'mean',
    'passenger_count': 'mean'
}).reset_index()

daily_trips.columns = ['date', 'total_trips', 'avg_duration', 'avg_passengers']
daily_trips['date'] = pd.to_datetime(daily_trips['date'])

# Visualizar tendencia
plt.figure(figsize=(15, 12))

# Tendencia de viajes
plt.subplot(3, 1, 1)
plt.plot(daily_trips['date'], daily_trips['total_trips'], linewidth=1, alpha=0.7, color='cornflowerblue')
plt.title('Tendencia Diaria de Viajes')
plt.xlabel('Fecha')
plt.ylabel('Número de Viajes')
plt.grid(True, alpha=0.3)

# Tendencia de duración
plt.subplot(3, 1, 2)
plt.plot(daily_trips['date'], daily_trips['avg_duration'], linewidth=1, alpha=0.7, color='mediumslateblue')
plt.title('Tendencia de Duración Promedio')
plt.xlabel('Fecha')
plt.ylabel('Duración Promedio (min)')
plt.grid(True, alpha=0.3)

# Tendencia de pasajeros
plt.subplot(3, 1, 3)
plt.plot(daily_trips['date'], daily_trips['avg_passengers'], linewidth=1, alpha=0.7, color='deeppink')
plt.title('Tendencia de Pasajeros Promedio')
plt.xlabel('Fecha')
plt.ylabel('Pasajeros Promedio')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Análisis de estacionalidad

```python
try:
    from statsmodels.tsa.seasonal import seasonal_decompose
    
    # Preparar datos para descomposición estacional
    ts_data = daily_trips.set_index('date')['total_trips']
    ts_data = ts_data.asfreq('D').fillna(method='ffill')
    
    # Descomposición estacional (usando 7 días como período para patrón semanal)
    decomposition = seasonal_decompose(ts_data, model='additive', period=7)
    
    # Visualizar descomposición
    plt.figure(figsize=(15, 12))
    
    plt.subplot(4, 1, 1)
    plt.plot(decomposition.observed)
    plt.title('Serie Original - Viajes Diarios')
    plt.ylabel('Viajes')
    plt.grid(True, alpha=0.3)
    
    plt.subplot(4, 1, 2)
    plt.plot(decomposition.trend)
    plt.title('Tendencia')
    plt.ylabel('Viajes')
    plt.grid(True, alpha=0.3)
    
    plt.subplot(4, 1, 3)
    plt.plot(decomposition.seasonal)
    plt.title('Estacionalidad (Patrón Semanal)')
    plt.ylabel('Viajes')
    plt.grid(True, alpha=0.3)
    
    plt.subplot(4, 1, 4)
    plt.plot(decomposition.resid)
    plt.title('Residuales')
    plt.ylabel('Viajes')
    plt.xlabel('Fecha')
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
except ImportError:
    print("statsmodels no disponible para descomposición estacional")
```

## Análisis de patrones por tipo de día

```python
# Comparar patrones horarios entre días laborables y fines de semana
weekday_hourly = df_final[df_final['is_weekend'] == False].groupby('pickup_hour').size()
weekend_hourly = df_final[df_final['is_weekend'] == True].groupby('pickup_hour').size()

# Duración promedio por tipo de día
weekday_duration_hourly = df_final[df_final['is_weekend'] == False].groupby('pickup_hour')['trip_duration_min'].mean()
weekend_duration_hourly = df_final[df_final['is_weekend'] == True].groupby('pickup_hour')['trip_duration_min'].mean()

plt.figure(figsize=(15, 8))

plt.subplot(1, 2, 1)
plt.plot(weekday_hourly.index, weekday_hourly.values, label='Días Laborables', linewidth=2, color='royalblue')
plt.plot(weekend_hourly.index, weekend_hourly.values, label='Fin de Semana', linewidth=2, color='deeppink')
plt.title('Patrón Horario: Laborables vs Fin de Semana')
plt.xlabel('Hora del Día')
plt.ylabel('Número de Viajes')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xticks(range(0, 24, 2))

plt.subplot(1, 2, 2)
plt.plot(weekday_duration_hourly.index, weekday_duration_hourly.values, label='Días Laborables', linewidth=2, color='royalblue')
plt.plot(weekend_duration_hourly.index, weekend_duration_hourly.values, label='Fin de Semana', linewidth=2, color='deeppink')
plt.title('Duración Promedio: Laborables vs Fin de Semana')
plt.xlabel('Hora del Día')
plt.ylabel('Duración Promedio (min)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xticks(range(0, 24, 2))

plt.tight_layout()
plt.show()
```

## Análisis de correlaciones temporales

```python
# Correlación entre variables diarias
correlation_data = daily_trips[['total_trips', 'avg_duration', 'avg_passengers']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_data, annot=True, cmap='bwr', center=0, fmt='.3f')
plt.title('Matriz de Correlación - Variables Diarias')
plt.show()

print("Matriz de correlación:")
print(correlation_data)
```

# 5. Resultados

```python
# Métricas resumen temporales
print(f" Período analizado: {df_final['pickup_datetime'].min().strftime('%Y-%m-%d')} a {df_final['pickup_datetime'].max().strftime('%Y-%m-%d')}")
print(f"  Total de días analizados: {daily_trips.shape[0]}")
print(f"  Promedio diario de viajes: {daily_trips['total_trips'].mean():.0f}")
print(f"  Desviación estándar diaria: {daily_trips['total_trips'].std():.0f}")
print(f"  Rango diario: {daily_trips['total_trips'].min():.0f} - {daily_trips['total_trips'].max():.0f} viajes")

print("\n Patrones horarios")
print(f"  Hora pico: {hourly_trips.idxmax()}:00 hrs ({hourly_trips.max():,.0f} viajes)")
print(f"  Hora valle: {hourly_trips.idxmin()}:00 hrs ({hourly_trips.min():,.0f} viajes)")
print(f"  Ratio pico valle: {hourly_trips.max()/hourly_trips.min():.1f}x")

print("\n Ptrones semanales")
print(f"  Día con más viajes: {weekday_names[weekday_trips.idxmax()]} ({weekday_trips.max():,.0f} viajes)")
print(f"  Día con menos viajes: {weekday_names[weekday_trips.idxmin()]} ({weekday_trips.min():,.0f} viajes)")
print(f"  Diferencia semanal: {((weekday_trips.max()/weekday_trips.min())-1)*100:.1f}%")

print("\n Comparativa fin de semana vs laborables")
weekday_avg_trips = df_final[df_final['is_weekend'] == False].shape[0] / (daily_trips.shape[0] * 5/7)
weekend_avg_trips = df_final[df_final['is_weekend'] == True].shape[0] / (daily_trips.shape[0] * 2/7)
weekday_avg_duration = df_final[df_final['is_weekend'] == False]['trip_duration_min'].mean()
weekend_avg_duration = df_final[df_final['is_weekend'] == True]['trip_duration_min'].mean()

print(f"  Viajes promedio días laborables: {weekday_avg_trips:.0f}/día")
print(f"  Viajes promedio fin de semana: {weekend_avg_trips:.0f}/día")
print(f"  Duración promedio días laborables: {weekday_avg_duration:.1f} min")
print(f"  Duración promedio fin de semana: {weekend_avg_duration:.1f} min")

print("\n Estadísticas generales")
print(f"  Duración promedio total: {df_final['trip_duration_min'].mean():.1f} min")
print(f"  Pasajeros promedio total: {df_final['passenger_count'].mean():.2f}")
print(f"  Total de viajes analizados: {len(df_final):,}")
```

## Conclusión

El análisis de los datos de viajes de taxis de NYC durante el primer semestre de 2016 revela patrones operativos claros y cuantificables. El período analizado comprendió 182 días con un promedio estable de 8,011 viajes diarios, demostrando una demanda consistente en el servicio de taxis.

Los patrones temporales identificados muestran una marcada variación en la demanda:
- **Patrón horario**: Se observa un ratio pico/valle de 6.0x, con la máxima demanda a las 18:00 hrs (90,559 viajes) y la mínima a las 5:00 hrs (14,984 viajes)
- **Patrón semanal**: El viernes registra la mayor actividad (223,414 viajes) mientras el lunes presenta la menor (187,333 viajes), con una diferencia del 19.3%
- **Comparativa laborables vs fin de semana**: Aunque la cantidad de viajes es similar, los días laborables presentan viajes ligeramente más largos (16.2 min vs 15.4 min)

El análisis de 1,457,949 viajes confirma que el servicio opera con un promedio de 1.66 pasajeros por viaje y una duración promedio de 16.0 minutos. Estos hallazgos proporcionan una base cuantitativa sólida para la optimización de flotas, distribución de recursos y planificación de horarios, permitiendo una asignación más eficiente basada en los patrones de demanda reales identificados.
